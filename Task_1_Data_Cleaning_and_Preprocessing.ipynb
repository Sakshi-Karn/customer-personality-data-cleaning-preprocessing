```python
import pandas as pd

df=pd.read_csv("marketing_campaign.csv",sep='\t')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Handling missing values
#Since Income has missing values
df['Income']=df['Income'].fillna(df['Income'].median())
```

```python
#dropping rows where date is missing
df=df.dropna(subset=['Dt_Customer'])
```

```python
#Remove duplicates
df.duplicated().sum()

df=df.drop_duplicates()
```

```python
#Fixing DATE Column
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'],dayfirst=True)
```

```python
# Rename Columns
df.columns = df.columns.str.lower()
df.columns

```

```python
#Standardize text columns
df['education'] = df['education'].str.lower().str.strip().replace({
    '2n cycle': 'second_cycle',
    'basic': 'basic',
    'graduation': 'graduation',
    'masters': 'master',
    'phd': 'phd'
})

df['marital_status'] = df['marital_status'].str.lower().str.strip().replace({
    'married': 'married',
    'together': 'together',
    'single': 'single',
    'divorced': 'divorced',
    'widow': 'widow',
    'alone': 'single'
})
```

```python
#fix data types
auto_numeric_cols = [
    col for col in df.columns
    if col not in ['education','marital_status','dt_customer']
]

for col in auto_numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#final null check
print(df.isnull().sum())
```

```python
#exporting cleaned data
df.to_csv("cleaned_marketing_campaign.csv",index=False)
```
